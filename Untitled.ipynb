We constant that there is a few number of columns that beholds string values instead of numerical values . Thus there are *some categorical variables to handle*

Furthermore, analysing the percentage of null values in each column + a little bit of scientific understanding of each column meaning makes us aware of the existance of categories we can use later while preprocessing while feature engineering the data

```python
#we already can drop the columns that have no value in them
#no value => no insight
cols_with_no_val_to_drop = [
    'Prothrombin time (PT), Activity',
    'D-Dimer',
    'Mycoplasma pneumoniae',
    'Urine - Sugar',
    'Partial thromboplastin time\xa0(PTT)\xa0'
]
df=df.drop(cols_with_no_val_to_drop, axis=1)
df
```

### Core analysis
Visualize and understand the relations between the variables and between some variables and the target column `'SARS-Cov-2 exam result'`

```python
pneumo_infections = df.columns[
    (df.isna().sum() / df.shape[0] > 0) &
    (df.isna().sum() / df.shape[0] <= 0.854713)
]
#here we have selected the columns that represent the pneumonic viral and bacterial infections , including the 2 rapid tests of influenza
pneumo_infections
```
